Given a string of round, curly, and square open and closing brackets, return whether the brackets are balanced (well-formed).

For example, given the string "([])[]({})", you should return true.

{}

Given the string "([)]" or "((()", you should return false.

"()[]{}", True

- This case is not correct: 

"({)}"

Solución viene de https://www.geeksforgeeks.org/check-for-balanced-parentheses-in-python/. Se hace entrega del ejercicio para dar constancia de que se ha entendido la solución empleada.

```python

open_list = ["[","{","("]
close_list = ["]","}",")"]
  
def balanced(string):
    stack = []
    for i in string:
        if i in open_list:
            stack.append(i)
        elif i in close_list:
            pos = close_list.index(i)
            if ((len(stack) > 0) and
                (open_list[pos] == stack[len(stack)-1])):
                stack.pop()
            else:
                return "Unbalanced"
    if len(stack) == 0:
        return "Balanced"
    else:
        return "Unbalanced"
  
```

```python
balanced("((([])))")
```

```python
balanced("[[({})]]")
```

```python
balanced("[[({}())])]")
```
