### 1. Se pide:

A partir del archivo "USA_Housing.csv", realizar la regresión no lineal entre cada columna numérica con la columna 'price' utilizando:

- Regresión polinómica no lineal con grados del 2 al 10. ¿El score se estanca a partir de cierto grado?

```python
# Import function to create training and test set splits
from sklearn.model_selection import train_test_split
# Import function to automatically create polynomial features! 
from sklearn.preprocessing import PolynomialFeatures
# Import Linear Regression
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold, RepeatedKFold

from sklearn.metrics import r2_score

import numpy as np
```

```python
import pandas as pd

df = pd.read_csv("USA_Housing.csv")
df
```

```python
lista_X = df[["Avg. Area Income", "Avg. Area House Age", "Avg. Area Number of Rooms", "Area Population", "Avg. Area Number of Bedrooms"]].values
l_X = np.transpose(lista_X)
l_X.shape
```

```python
y = df["Price"].values
y.shape
```

```python
lin_reg_model = LinearRegression()
LL = ["Avg. Area Income", "Avg. Area House Age", "Avg. Area Number of Rooms", "Area Population", "Avg. Area Number of Bedrooms"]

for i, X in enumerate(l_X):
    
    X = X.reshape(-1, 1)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    print("\n")
    tabla_score = []
    tabla_x = []
    for degree in range(2,11):
        polinominal_model = PolynomialFeatures(degree) 
        X_poly = polinominal_model.fit_transform(X_train.reshape(-1, 1), y_train)
        lin_reg_model.fit(X_poly, y_train)
        

        y_pred = lin_reg_model.predict(X_poly)
        tabla_score.append(r2_score(y_train, y_pred)*100)
        tabla_x.append(degree)
        print (f'Score con "{degree}" grados de columna {LL[i]} para Price:', r2_score(y_train, y_pred)*100)
    
    print("\n")
    plt.plot(tabla_x, tabla_score)
    plt.title("Estancamiento de scores para grados")
    plt.xlabel("Grados")
    plt.ylabel("Score")
    plt.show()
    
```

#### Avg. Area Income se estanca a partir de los 5 grados
#### Avg. Area House Age no llega a estancarse
#### Avg. Area Number of Rooms no llega a estancarse
#### Area Population se estanca a partir de los 3 grados
#### Avg. Area Number of Bedrooms no llega a estancarse
