### Step 1. Import the necessary libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# remember to %matplotlib inline
%matplotlib inline
```

### Step 2. Import the dataset of `chipotle` and assign it to a variable called chipo.

```python
chipo = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", sep='\t')
chipo
```

### Step 3. See the first 10 entries

```python
chipo.iloc[0:10]
```

### Step 4. Create a histogram of the top 5 items bought

```python
top = chipo.groupby("item_name").sum().sort_values(by="quantity", ascending=False)
top = top.reset_index()
top = top.drop(columns="order_id")

top5 = top.iloc[0:5].sort_values(by="quantity", ascending=True)

```

```python
plt.plot(top5.item_name, top5.quantity)

plt.show()
```

```python
plt.bar(top5["item_name"],top5["quantity"], width=0.5)
plt.ylabel('Number of times ordered')
plt.xlabel('Items')
plt.xticks(ticks = top5["item_name"], labels = (top5["item_name"]), rotation=90)
plt.title("Most ordered Chipotle's Items")
plt.show()
```

### Step 5. Create a scatterplot with the number of items orderered per order price

Make sure you get the same labels and title
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
chipo['item_price'] = chipo['item_price'].str[1:].astype(float)
chipo_tickets = chipo.groupby("order_id").sum()
chipo_tickets
```

```python
plt.scatter(x=chipo_tickets["item_price"],y=chipo_tickets["quantity"], color="green")
plt.title("Number of items ordered per order price")
plt.xlabel('Order Price')
plt.ylabel('Items ordered')
```
