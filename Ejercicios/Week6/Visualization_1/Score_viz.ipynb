### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# set seaborn graphs to a better style
sns.set(style="ticks")
```

### Step 2. Create the DataFrame it should look like below.

This time you will create the data.

```python
dictt = {'first_name': ["Jason","Molly", "Tina", "Jake", "Amy"], 'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', "Cooze"],'age': [42, 52, 36, 24, 73],'female': [0, 1, 1, 0, 1],"preTestScore": [4, 24, 31, 2, 3],'postTestScore': [25, 94, 57, 62, 70]}
data = pd.DataFrame(dictt)
data
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
sns.scatterplot(size = "age", x = "preTestScore", y = "postTestScore", data = data, legend=False, palette="blues")
plt.title("Number of items ordered per order price")
plt.xlabel("preTestScore")
plt.ylabel("postTestScore")
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python

plt.scatter(x="preTestScore", y = "postTestScore", data = data, s="age", linewidths=6)
plt.title("Number of items ordered per order price")
plt.xlabel("preTestScore")
plt.ylabel("postTestScore")
```
